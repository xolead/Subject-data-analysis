```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
"""
1 Указание: рекомендуется воспользоваться функцией genfromtxt. Обратите внимание, что данные в файлах разделены запятой, в обоих файлах есть заголовок, 
который при загрузке нужно пропустить. Для корректной загрузки названий стран необходимо также указать тип данных 
(по умолчанию загружаемые данные преобразуются в вещественный тип). Вы можете либо собрать все данные в один массив записей,
 либо работать с тремя разными массивами: массив названий стран (одномерный, строки), массивы производства и потребления энергии (двумерные, вещественные)).
"""
countries = np.genfromtxt(
    "global-electricity-consumption.csv",
    delimiter=',', 
    skip_header=1,
    usecols=0,  
    dtype=str,
    encoding='utf-8'
)

consumption_values = np.genfromtxt(
    "global-electricity-consumption.csv",
    delimiter=',', 
    skip_header=1,
    usecols=range(1, 31),
    dtype=float,
    encoding='utf-8'
)

generation_values = np.genfromtxt(
    "global-electricity-generation.csv",
    delimiter=',', 
    skip_header=1,
    usecols=range(1, 31),
    dtype=float,
    encoding='utf-8'
)

print("Страна:", countries[0])
print("Потребление:", consumption_values[0])
print("Производство:", generation_values[0])
```

```python
consumption_last_5 = consumption_values[:, -5:]
print("последние 5 записей потребления электроэнергии по всем странам")
print("-" * 62 )
print(consumption_last_5)
generation_last_5 = generation_values[:, -5:]
```

```python
"""
2 Постройте одномерные массивы ежегодного производства и потребления электроэнергии в среднем за последние 5 лет 
(один массив – производство по всем странам, второй – потребление по всем странам).
Указание: не забывайте, что в агрегирующих функциях для построчной обработки нужно указывать параметр axis=1.
"""
average_generation_5_last = np.nanmean(generation_last_5, axis=1)

average_consumption_5_last = np.nanmean(consumption_last_5, axis=1)

print(f"{'Страна':<25} {'Потребление':<12} | {'Страна':<25} {'Производство':<12}")
print("-" * 39 + "+" + "-" * 39)
for i in range(len(countries)):
    print(f"{countries[i]:<25} {average_consumption_5_last[i]:12.2f} | {countries[i]:<25} {average_generation_5_last[i]:12.2f}")
```

```python
"""
3.1 Суммарное (по всем странам) потребление электроэнергии за каждый год.
"""
total_consumption_by_year = np.nansum(consumption_values, axis=0)

years = np.arange(1992, 2022)
print(f"\n{'Год':<6} {'Потребление':<20}")
print("-" * 18)
for i, year in enumerate(years):
    print(f"{year:<6} {total_consumption_by_year[i]:<20.2f}")
```

```python
"""
3.2 Максимальное количество электроэнергии, которое произвела одна страна за один год 
(указание: чтобы не учитывать отсутствующие и некорректные данные (nan) 
воспользуйтесь NaN-безопасной версией функции max, то есть nanmax).
"""
maximum_generation = np.nanmax(generation_values, axis=0)

max_country_indices = np.nanargmax(generation_values, axis=0)

years = np.arange(1992, 2022)
print(f"\n{'Год':<6} {'Макс. выработка':<16} {'Страна':<20}")
print("-" * 50)
for i, year in enumerate(years):
    country_idx = max_country_indices[i]
    country_name = countries[country_idx]
    max_value = maximum_generation[i]
    print(f"{year:<6} {max_value:<16.2f} {country_name:<20}")
```

```python
"""
3.3 Список стран, которые производят более 500 млрд. кВт*ч электроэнергии ежегодно в среднем за последние 5 лет (воспользуйтесь массивом, полученным на шаге 2).
"""
high_production_countries_index = np.where(average_generation_5_last > 500)[0]

print(f"Страны, производящие более {500} ТВт·ч ({500} млрд. кВт·ч) в среднем за 2017-2021 годы:")
print("=" * 70)
print(f"{'Страна':<25} {'Среднее производство (ТВт·ч)':<25}")
print("-" * 70)

for i in high_production_countries_index:
    print(f"{countries[i]:<25} {average_generation_5_last[i]:<25.2f}")
```

```python
"""
3.4. 10% стран, которые потребляют больше всего электроэнергии ежегодно в среднем за последние 5 лет 
(указание: вначале определите соответствующую квантиль в массиве, построенном на шаге 2).
"""

n_countries = len(countries)
top_10_percent = int(np.ceil(n_countries * 0.1))

sorted_indices_consu = np.argsort(-average_consumption_5_last)

top_indices = sorted_indices_consu[:top_10_percent]

print(f"Топ {top_10_percent} стран по потреблению электроэнергии")
print("=" * 70)
print(f"{'Страна':<25} {'Среднее потребление (ТВт·ч)':<25}")
print("-" * 70)

for i in top_indices:
    country = countries[i]
    consumption = average_consumption_5_last[i]
    print(f"{country:<25} {consumption:<25.2f}")
```

```python
"""
3.5. Список стран, которые увеличили производство электроэнергии в 2021 году по сравнению с 1992 годом более, чем в 10 раз.
"""
all_in_1992 = generation_values[:, 0]
all_in_2021 = generation_values[:, -1]

# Избегаем деления на ноль, np.where(условие, значение_если_да, значение_если_нет)
all_in_1992_safe = np.where(all_in_1992 == 0, 0.001, all_in_1992)
growth_ratio = all_in_2021 / all_in_1992_safe

high_growth_indices = np.where(growth_ratio > 10)[0]

print(f"Страны, увеличившие производство электроэнергии более чем в 10 раз (2021 vs 1992)")
print("=" * 90)
print(f"{'Страна':<25} {'1992 год (ТВт·ч)':<15} {'2021 год (ТВт·ч)':<15} {'Рост (раз)':<12}")
print("-" * 90)

for i in high_growth_indices:
    print(f"{countries[i]:<25} {all_in_1992[i]:<15.2f} {all_in_2021[i]:<15.2f} {growth_ratio[i]:<12.1f}")
```

```python
"""
3.6. Список стран, которые в сумме за все годы потратили больше 100 млрд. кВт*ч электроэнергии и при этом произвели меньше, чем потратили.
"""

total_consumption_by_all = np.nansum(consumption_values, axis = 1)
total_generation_by_all = np.nansum(generation_values, axis = 1)

idx_spent_more_100_and_gen_below = np.where((total_consumption_by_all > 100) & (total_consumption_by_all > total_generation_by_all))[0]

print(f"Страны с потреблением >100 ТВт·ч и дефицитом производства")
print("=" * 80)
print(f"{'Страна':<25} {'Потребление':<12} {'Производство':<12}")
print("-" * 80)

for i in idx_spent_more_100_and_gen_below:
    print(f"{countries[i]:<25} {total_consumption_by_all[i]:<12.2f} {total_generation_by_all[i]:<12.2f}")

print(f"\nВсего стран: {len(idx_spent_more_100_and_gen_below)}")
```

```python
"""
3.7. Какая страна потратила наибольшее количество электроэнергии в 2020 году?
"""

consumption_2020 = consumption_values[:, -2]

country_max_spent_2020 = np.argmax(consumption_2020)

print(f"Страна с максимальным потреблением в 2020 году: {countries[country_max_spent_2020]}")
print(f"Потребление: {consumption_2020[country_max_spent_2020]:.2f} ТВт·ч")
```
