```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
"""
Сгенерировать два одномерных массива m1 и m2 размерностью 10 (значения от 1 до 10).
"""
m1 = np.random.randint(1, 11, 10)
m2 = np.random.randint(1, 11, 10)

print("m1 = ", m1)
print("m1 = ", m2)
```

```python
"""
1. Сформировать на их основе массивm3, содержащий только те элементы, которые есть либо только в массиве m1, либо только в массиве m2.
"""

m3 = np.setxor1d(m1, m2)

print("m3 :", m3)
```

```python
"""
2. Заменить в первом массиве все значения, кратные 3 или 2 на число 1, а остальные – оставить без изменений.
"""

replaced_m1 = np.where((m1 % 3 == 0) | (m1 % 2 == 0), 1, m1)
print("После замены:", replaced_m1)
```

```python
"""
3. Слить массивыm1 и m2 в один и преобразовать получившийся одномерный массив в матрицу размерностью 4х5. Вывести ее на экран.
"""

merged = np.concatenate([m1, m2])
print("\nОбъединенный массив:", merged)

matrix_4x5 = merged.reshape(4, 5)
print("\nМатрица 4×5:")
print(matrix_4x5)
```

```python
"""
4. Удалить из матрицы 1 и 4 столбцы.
"""

#matrix_without_cols = matrix_4x5[:, [1, 2, 4]]
matrix_without_cols = np.delete(matrix_4x5, [0, 3], axis=1)
print("\nПосле удаления 1 и 4 столбцов (3×3):")
print(matrix_without_cols)
```

```python
"""
5. Транспонировать полученную матрицу.
"""

transposed_matrix = np.transpose(matrix_without_cols)
print(transposed_matrix)
```
